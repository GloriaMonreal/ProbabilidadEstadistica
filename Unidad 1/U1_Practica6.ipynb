```python
#Importamos todas las librerias a usar
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
from statistics import *
import math
import numpy as np
```

##1.Cargar el Dataset de Autos y realizar las Estadísticas descriptivas básicas y medidas de tendencia central

```python
#√alor maximo
autos.price.max()
```

```python
#√alor minimo
autos.price.min()
```

```python
#Moda - mode
moda = mode(autos.price)
print("La moda de precios es: ",moda)
```

```python
#Moda - multimode
moda1 = multimode(autos.price)
print("La moda de precios es: ",moda)
```

```python
#Media
media = np.mean(autos.price)
print("La media de precios es: ",media)
```

```python
#Mediana
mediana = median(autos.price)
print(mediana)
```

```python
#Desviacion estandar
desv_estandar_precios = np.std(autos['price'])  # Use NumPy's std function
print(desv_estandar_precios)
```

```python
#Estadísticos y cuartiles
autos.price.describe()
```

##2. Realizar un Histograma de precios de los automóviles

```python
#Regla de Sturges
n = len(autos.price)
k = int(1 + math.log2(n))
print("Valor de n: ", n, " \nNumero de intervalos: ", k)
```

```python
#Histograma
plt.pyplot.hist(autos.price, bins = 8, alpha = 0.2, color = "b", edgecolor = "black")
plt.pyplot.axvline(media, color = "b", linestyle = "dashed", linewidth = 2, label = f"Media: {media:.2f}")
plt.pyplot.axvline(mediana, color = "r", linestyle = "dashed", linewidth = 2, label = f"Mediana: {mediana:.2f}")
plt.pyplot.axvline(moda, color = "y", linestyle = "dashed", linewidth = 2, label = f"Moda: {moda:.2f}")
plt.pyplot.axvline(media - desv_estandar_precios,color = "black",linestyle = "dashed",linewidth = 2,label = f"Desv.Estandar: {desv_estandar_precios:.2f}")
plt.pyplot.axvline(media + desv_estandar_precios,color ="black",linestyle ="dashed",linewidth =2)
plt.pyplot.xlabel("Precio")
plt.pyplot.ylabel("Frecuencia")
plt.pyplot.title("Precio de los Autos")
plt.pyplot.legend()
plt.pyplot.show()
```

**Conclusiones**

En este histograma podemos visualizar la relacion entre precio y la frecuencia en la que podemos encontrar dicho precio en el documento de autos2.cvs, y somos capaces de apreciar todas las medidas de tendencia central que se agregaron; al seguir la linea azul punteada podemos observar la media con un valor de 13,207.13, al encontrar la linea roja punteada encontramos la mediana con un valor de 18,150.00, y al buscar la linea punteada de color amarilla podemos apreciar el valor de la moda, dandonos cuenta que esta tiene un valor de 16,500.00; de igual manera, somos capaces de apreciar la desviacion estandar al encontrar las lineas punteadas de color negro, esta contando con un valor de 7,927.27; a partir de este histograma somos capaces de apreciar datos de suma importancia y relevantes de la materia, mismos datos que hemos venido trabajando a lo largo de la unidad.

##3. Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y tipo de vehículo (body style)

```python
#diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y tipo de vehículo (body style)
sns.boxplot(x = "body-style",y = "price", data = autos)
plt.pyplot.xlabel("Tipo de automovil")
plt.pyplot.ylabel("Precio del automovil")
plt.pyplot.title("Relacion Tipo-Precio de los Automoviles")
```

**Conclusiones**
El diagrama de cajas (boxplot) representa la relación entre el tipo de automóvil y su precio. Aquí hay algunos puntos clave de interpretación.Al observalo detenidamente podemos ver que los convertibles y hardtops tienen los precios más altos en general, ya que cuentan con medianas alrededor de 15,000 y 20,000 respectivamente, mientras que los hatchbacks, sedanes y wagons tienen precios visiblemente más bajos, con medianas cercanas a los 10,000 aproximadamente.

Esto nos dice que hay una diferencia muy clara en los precios basandonos en el tipo de vehículo. Es posible que los convertibles y hardtop sean autos de alta gama o de lujo, mientras que los hatchbacks, sedanes y vagones sean marcas de vehiculos de gama media o baja, teniendo por ende precios mas accesibles.También es importante mensionar que los precios dentro de cada categoría puede depender de mas factores como modelo, año de fabricación y características adicionales, lo que explica a su vez la presencia de algunos valores atípicos en las categorias de los hatchbacks, sedanes y vagones.

##4. Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y tipo tracción ruedas (drive wheels)

```python
#diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y tipo tracción ruedas (drive wheels)
sns.boxplot(x = "drive-wheels",y = "price", data = autos)
plt.pyplot.xlabel("Tipo tracción de las ruedas")
plt.pyplot.ylabel("Precio del automovil")
plt.pyplot.title("Relacion Traccion-Precio de los Automoviles")
```

**Conclusiones**

Este diagrama de cajas o bigotes muestra la relación entre el tipo de tracción de las ruedas y el precio de los automóviles.De primera impresion podemos ver que los vehículos con tracción trasera (rwd) tienden a tener precios más altos, con una mediana por arriba de los 15,000 y un rango un tanto extenso,igualmente apreciamos que esta categoria incluye algunos valores atípicos que superan los 35,000. Por otro lado, los vehículos con tracción delantera (fwd) cuentan con la mediana más baja de todas, alrededor de los 10,000, con un rango menor que los vehiculos con traccion trasera, aunque también hay varios valores atípicos que se encuentran por ensima de los 15,000. De igual manera, los vehículos con tracción en las cuatro ruedas (4wd) tienen precios intermedios aunque mucho mas cercanos a los vehiculos con tracccion delantera, ya que cuentan con una mediana similar a la de los fwd, pero tienen un rango visiblemente menor y solo se aprecia un valor atípico.

 A partir de este diagrama podemos deducir que en efecto existe una diferencia significativa al considerar el tipo de traccion del vehiculo para calcular el precio, en general las tracciones traseras son principalmente utilizadas en autos deportivos, de lujo y de alto rendimiento, lo que podria explicar los precio tan altos, la traccion en las cuatro ruedas suele ser usado en vehículos todoterreno o similares, por lo que su precio no varia mucho, lo que explicaria el rango tan pequeño de precios, y por ultimo, el sistema de tracción delantero es más común en vehículos de gama media y baja, lo que justifica su menor precio. Aunque es importante mencionar que pueden influir mas factores como modelo o algunas caracteristicas adicionales, lo que explica la aparicion de los valores atipicos en cada una de las categorias.

##5. Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y la ubicación del motor (engine-location)

```python
#diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y la ubicación del motor (engine-location)
sns.boxplot(x = "engine-location",y = "price", data = autos)
plt.pyplot.xlabel("Ubicación del motor")
plt.pyplot.ylabel("Precio del automovil")
plt.pyplot.title("Relacion Ubicacion-Precio de los Autos")
```

**Conclusiones**

Al observar el diagrama de cajas, se aprecia que los vehículos con motor ubicado al frente (front) muestarn precios en un rango más bajo y con una mediana menor que los vehículos con motor trasero (rear). En la categoría “front” se ve una mayor cantidad de precios teniendo en cuenta valores mas pequeños, aunque tiene algunos valores atípicos que se elevan. Por el contrario, los autos con motor trasero muestran precios más altos, con una mediana visiblemente mayor y aparentemente no cuenta con valores atipicos.

Esta diferencia de precios podría deberse a muchos factores, como el tipo de vehículo o su propósito. Los autos con motor trasero suelen ser modelos deportivos o de lujo, lo que podria explicr su alto costo. En cambio, los vehículos con un motor en la parte de enfrente son más comunes, e incluyen desde autos económicos hasta de gama media, lo que podria explicar el mayor rango de precios, y aunque existen algunos valores atípicos en esta categoría, la mayoría de estos vehículos tienen precios relativamente bajos.
