### **Problema 1**
Realizar un programa en Python que resuelva el siguiente problema:

• El número x de personas ingresadas a una unidad de cuidados intensivos en un hospital particular, en un día, tiene una distribución de probabilidad de Poisson con media igual a cinco personas por día.

• a). ¿Cuál es la probabilidad de que el número de personas ingresadas a una unidad de cuidados intensivos en un hospital particular, en un día particular, sea dos?

• b). ¿Es probable que x exceda de 10? Explicar

-----------------------------------------
Este problema se trata de un problema de Distribucion de Poisson ya que, a pesar de que en el mismo problema menciona que se "tiene una distribucion de probabilidad de Poisson", por lo que puedo deducir que se trata de una distribucion de poisson, si nos ponemos a analizarlo un poco podemos encontrar:

*   El número de veces que ocurre un evento en los ingresos de la Unidad de Cuidados Intencivos
*   Un intervalo de tiempo fijo de 1 día.
*   La media o tasa esperada (λ = 5).
*   Los eventos son independientes.

Toda esta informacion forma parte de la Distribucion de Poisson






```python
import numpy as np
from scipy.stats import poisson
```

```python
# Parámetro lambda (media)
media = 5

# a) Probabilidad de que ingresen 2 personas en un día
prob_2 = poisson.pmf(2, media)
print(f"a) La probabilidad de que ingresen 2 personas en un día es: {prob_2:.4f}")

# b) Probabilidad de que ingresen más de 10 personas en un día
prob_mayor_10 = poisson.sf(10, media)  # sf = 1 - cdf
print(f"b) La probabilidad de que ingresen más de 10 personas en un día es: {prob_mayor_10:.4f}")

# Explicación adicional
if prob_mayor_10 < 0.05:
    print("→ No es muy probable que ingresen más de 10 personas en un día (menos del 5%).")
else:
    print("→ Es bastante probable que ingresen más de 10 personas en un día.")
```

## **Problema 2**
Realizar un programa en python que resuelva el siguiente problema:

Un fabricante de podadoras para el pasto compra motores de 1 hp y 2 ciclos, en lotes de 1,000, a un proveedor. El fabricate equipa cada una de las podadoras producidas por su planta con uno de los motores. La historia muestra que la probabilidad de que cualquier motor del proveedor resulte no satisfactorio es .001. En un embarque de 1,000 motores.

**Calcular**

• ¿cuál es la probabilidad de que ninguno sea defectuoso?

• ¿cuál es la probabilidad de que tres sean defectuosos?

• ¿cuál es la probabilidad de que cuatro sean defectuoso?

-----------------------------------------
Este problema se trata de un problema de Distribucion de Binomial ya que, al analizar el problema un poco podemos encontrar:

*   Un número fijo de ensayos (n = 1000 motores).
*   Cada ensayo tiene dos posibles resultados: defectuoso o no defectuoso.
*   La probabilidad de éxito (defectuoso) es constante (p = 0.001).
*   Los ensayos son independientes.

Toda esta informacion son caracteristicas y datos que forma parte de la Distribucion Binomial






```python
from scipy.stats import binom
```

```python
# a) Probabilidad de que ninguno sea defectuoso
# Parámetros del problema
n = 1000          # Tamaño del lote
p = 0.001         # Probabilidad de que un motor sea defectuoso

# Calcular la probabilidad de que ningún motor sea defectuoso
prob_no_defect = binom.pmf(0, n, p)
print('a)La probabilidad de que ningun motor sea defectuoso es: ', prob_no_defect)

# b) Probabilidad de que exactamente 3 sean defectuosos
prob_3_defect = binom.pmf(3, n, p)
print('b)La probabilidad de que exactamente 3 motores sean defectuoso es: ', prob_3_defect)

# c) Probabilidad de que exactamente 4 sean defectuosos
prob_4_defect = binom.pmf(4, n, p)
print('c)La probabilidad de que exactamente 4 motores sean defectuoso es: ', prob_4_defect)
```

## **Problema 3**
Una revisión de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviación estándar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

*   La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.
*   Cuantos estudiantes pesan más de 68 y menos de 72 kg.


-----------------------------------------
Este problema se trata de un problema de Distribucion Normal ya que, a pesar de que en el mismo problema menciona "Suponiendo que los valores se distribuyen normalmente", por lo que puedo deducir que se trata de una Distribucion Normal, si nos ponemos a analizarlo un poco podemos encontrar:

*   Una media y una desviación estándar.
*   Una variable continua (peso).

Toda esta informacion forma parte de la Distribucion Normal





```python
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Parametros del problema
media = 70  # Media
desv_std = 4  # Desviación estándar
estudiantes = 500  # Número total de estudiantes

# a) Probabilidad de que el peso esté entre 68 y 72 kg
prob_68_72 = norm.cdf(72, media, desv_std) - norm.cdf(68, media, desv_std)
print('a) La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg es de: ', prob_68_72)

# b) Número de estudiantes que pesan entre 68 y 72 kg
estudiantes_68_72 = int(prob_68_72 * estudiantes)
print(f"b) Aproximadamente {estudiantes_68_72} estudiantes pesan entre 68 y 72 kg.")
```

```python
# Graficar la distribución normal
# Rango de valores para graficar la curva normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

# Área entre 68 y 72 kg
x_fill = np.linspace(68, 72, 1000)
y_fill = norm.pdf(x_fill, media, desv_std)

plt.figure(figsize=(10,6))
plt.plot(x, y, label='Distribución Normal', color='b')
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.7, label=f'Área entre 68 y 72 kg\nProbabilidad ≈ {prob_68_72:.4f}')
plt.axvline(68, color='gray', linestyle='--')
plt.axvline(72, color='gray', linestyle='--')
plt.title("Distribución Normal del Peso de Estudiantes", fontsize=14)
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.legend()
plt.show()
```
