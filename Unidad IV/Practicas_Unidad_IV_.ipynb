# **Distribucion Normal**

### **Ejercicio 1**
Una revision de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviación estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente.

Hallar lo siguiente:

• Generar aleatoriamente las edades de los 500 estudiantes con una media de 70 kg. y una desviación estandar de 4 kg..

• Realizar un histogrma de las edades

• Graficar la función de densidad de probabilidad (PDF).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generar datos para la distribución normal
media = 70
# Media
desv_std = 4
# Desviación estándar
num_datos = 500
# generamos Los datos de manera aleatoria y lo ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos [0:50] # Mostrar Los primeros 50
```

```python
##Crear un rango de valores para crear una densidad
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)
#Calcular La PDF de la distribución normal
pdf = 1/ (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media) **2 / (2 * desv_std **2))
```

```python
#Para no imprimir toda la Lista de valores (500), desplegamos Los primeros 50
print("Imprimir los primeros 50 Valores de x", x[0:50], " , y los primeros 50 pdf", pdf[0:50])
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b')#Histograma de datos
plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
#Visualizar La PDF
plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label='Normal: (media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
plt.title('Función de Densidad de Probabilidad (PDF) - Distribución Normal')
plt.xlabel('Pesos en Kg. de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot(x, pdf, 'k', linewidth=2,)
plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

### **Ejercicio 2**
Una revision de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviacion estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente.

Hallar lo siguiente:

• Cuantos estudiantes pesan menos de 75 kg.

• Cuantos estudiantes pesan menos de 62 kg.

• Cuantos estudiantes pesan mas de 79 kg.

Primero necesitamos calcular la probabilidad de que un estudiante pese menos de 75 kg utilizando la función de distribución acumulativa (CDF) de la distribución normal

```python
#Parámetros de La distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
total_estudiantes = 500
#VARIABLE para Cálculo de la probabilidad de que un estudiante que pese menos de 75 kg
peso_limite = 79

#Primero necesitamos calcular la probabilidad de que un estudiante pese menos
#de 75 kg utilizando La función de distribución acumulativa (CDF) de la distribución normal
prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)

# Cálculo del número de estudiantes que pesan menos de 75 kg
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg

print("Probabilidad de que un estudiante pese menos de 75 kg:", prob_menos_75kg)
print("Número de estudiantes que pesan menos de 75 kg:", estudiantes_menos_75kg)
```

```python
plt.figure(figsize=(6,4))
# Graficar la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between (x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75 kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

Cuantos estudiantes pesan menos de 62 kg.

```python
# Parámetros de la distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
total_estudiantes = 500

# VARIABLE para Cálculo de la probabilidad de que un estudiante pese menos de 75 kg
peso_limite = 62

#Primero necesitamos calcular La probabilidad de que un estudiante pese menos
#de 75 kg utilizando la función de distribución acumulativa (CDF) de la distribución normal
prob_menos_62kg = norm.cdf(peso_limite, media, desv_std)

# Cálculo del número de estudiantes que pesan menos de 75 kg
estudiantes_menos_62kg = total_estudiantes * prob_menos_62kg

print("Probabilidad de que un estudiante pese menos de 62 kg:", prob_menos_62kg)
print("Número de estudiantes que pesan menos de 75 kg:", estudiantes_menos_62kg)

#Calcular la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)
```

```python
plt.figure(figsize=(6,4))
#Graficar la distribución normal
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between (x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 62 kg, \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

Cuantos estudiantes pesan mas de 79 kg.

```python
# Parámetros de La distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
peso_limite = 79 #Peso Limite
total_estudiantes = 50

# Calculamos La probabilidad de que un estudiante pese menos o igual a 79 kg
prob_menos_79kg = norm.cdf(peso_limite, media, desv_std)

# La probabilidad de que un estudiante pese más de 79 kg es complementaria
prob_mas_79kg = 1 - prob_menos_79kg
# Calculamos el número aproximado de estudiantes que pesan más de 79 kg
estudiantes_mas_79kg = total_estudiantes * prob_mas_79kg

print("Probabilidad de que un estudiante pese más de 79 kg:", prob_mas_79kg)
print("Número aproximado de estudiantes que pesan más de 79 kg:", estudiantes_mas_79kg)

# Calcular La función de densidad de probabilidad de la distribución normal
x = np.linspace (media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 79 Kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

# Distribucion Binomial

### **Ejercicio 1**

**Ejercicio 1.1.** La probabilidad de éxito de una vacuna es de 0.83.

Calcular la probabilidad de que una vez administrada a 16 pacientes NINGUNO sufra la enfermedad.

```python
from scipy.stats import binom
```

```python
# Parámetros de la distribución binomial
probabilidad_exito = 0.83
num_pacientes = 16
X = 16 # Numero de exitos

# Calcular la probabilidad de que ningún paciente sufra la enfermedad
probabilidad_no_enfermos = binom.pmf(X, num_pacientes, probabilidad_exito)
print("La probabilidad de que ningún paciente sufra la enfermedad es:", probabilidad_no_enfermos)
```

**Ejercicio 1.2.** La probabilidad de éxito de una vacuna es de 0.83.

Calcular la probabilidad de que una vez administrada a 16 pacientes TODOS sufra la enfermedad.

```python
# Parámetros de la distribución binomial
probabilidad_exito = 0.83
num_pacientes = 16
X = 0
# Calcular la probabilidad de que todos los pacientes sufran La enfermedad
probabilidad_todos_enfermos = binom.pmf (X, num_pacientes, probabilidad_exito)
print("La probabilidad de que todos los pacientes sufran la enfermedad es:", probabilidad_todos_enfermos)
```

# **Distribucion de Poisson**

### **Ejercicio 1**
**Ejercicio 1.1.** Encuentre la probabilidad de que no haya accidentes en este crucero de carretera durante un periodo de 1 semana.

```python
from scipy.stats import poisson
```

```python
#Parámetros de la distribución de Poisson
media_accidentes = 2

probabilidad_0_accidentes = poisson.pmf(0, media_accidentes)
print("Probabilidad de que no haya accidentes en 1 semana:", probabilidad_0_accidentes)
```

**Ejercicio 1.2.** Encuentre la probabilidad de que a lo sumo haya tres accidentes en esta sección de carretera durante un periodo de 2 semanas.

```python
# Convertimos La media a una media para 2 semanas
media_2_semanas = media_accidentes * 2
#Calculamos la probabilidad acumulada hasta 3 accidentes en 2 semanas
probabilidad_3_o_menos_accidentes = poisson.cdf(3, media_2_semanas)
print("Probabilidad de que a lo sumo haya tres accidentes en 2 semanas:", probabilidad_3_o_menos_accidentes)
```
