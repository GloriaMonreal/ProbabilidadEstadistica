```python
from sklearn.linear_model import LinearRegression #Modelo de regresión Lineal
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Datos de entrada
periodos = [2018, 2019, 2020, 2021, 2022, 2023]
inversion = np.array([2, 3, 5, 4, 11, 5] ).reshape(-1, 1)
ganancias = [20, 25, 34, 30, 40, 31]

#Crear y entrenar el modelo de regresión Lineal
modelo = LinearRegression()
modelo.fit(inversion, ganancias)

#Predecir la ganancia para el año 2024
inversion_predict = np.array([[8]])
ganancia_predict = modelo.predict(inversion_predict)
print("La ganancia esperada para el año 2024 es de: ", ganancia_predict[0])

#Coeficientes de la regresión
print("Coeficiente de la pendiente (m):", modelo.coef_[0])
print("Intercepcion de la recta (b):", modelo.intercept_)
```

```python
 #Definir tamaño
plt.figure(figsize=(6, 4))

#Datos de entrada
inversiones = [2, 3, 5, 4, 11, 5]
ganancia = [20, 25, 34, 30, 40, 31]

#Crear el diagrama de dispersión con Línea de regresión
sns.set(style="whitegrid")
sns.regplot(x=inversiones, y=ganancia, scatter=True, color='blue')
plt.title('Diagrama de Dispersión con Línea de Regresión')
plt.xlabel('Inversión (millones)')
plt.ylabel('Ganancias (millones)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(inversion, ganancias, color = 'blue', label = 'Datos')
plt.plot(inversion, modelo.predict(inversion), color = 'red', label = 'Linea de regresión')
plt.scatter(inversion_predict, ganancia_predict, color ='green', marker = 'x', s=100, label ='Prediccion (8, {0:2f})'.format(ganancia_predict[0]))
plt.title('Pronostico de ganancias a partir de la inversión en publicidad')
plt.xlabel('Inversión en Publicidad (millones)')
plt.ylabel('Ganancias Anuales (millones)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import pearsonr
#Datos de entrada
publicidad= [2, 3, 5, 4, 11, 5]
ganancias = [20, 25, 34, 30, 40, 31]
#Calcular el coeficiente de correlación de Pearson
coeficiente_correlacion, p_valor = pearsonr(publicidad, ganancias)

print("Coeficiente de correlación de Pearson:", coeficiente_correlacion)
print("Valor p:", p_valor)
```
