````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
from scipy.stats import pearsonr
````

# **Grupo 1**

### **Carga y Exploración de Datos:**

````python
#Cargar el conjunto de datos del grupo 1 utilizando Python (pandas).
df1=pd.read_csv('/content/grupo1.csv')
df1.head()
````

````python
#Realizar una exploración inicial de los datos (descripción estadística y visualización).
df1.info()
````

````python
df1.describe()
````

````python
#Variable independientes o variables predictorias
x1 = df1 [["HorasEstudio"]]
x1.head(10)
````

````python
#variable dependiente o variable objetiva a predecir
y1 = df1 [["Calificacion"]]
y1.head(10)
````

````python
#Generar un histograma de calificaciones
plt.figure(figsize=(6, 4))
sns.histplot(df1["Calificacion"], kde=True, bins=10, color='blue', edgecolor='black')
plt.title('Grupo 1 - Histograma de Calificaciones')
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
````

### **Modelo de Regresión Lineal:**

````python
#Entrenar un modelo de regresión lineal simple usando scikit-learn o statsmodels.
lm1 = LinearRegression()
lm1.fit(x1, y1)
````

````python
#Obtener la ecuación de la línea de regresión (pendiente e intercepto).
print("Coeficiente de la pendiente (m):", lm1.coef_[0])#Pendiente
print("Intercepcion de la recta (b):", lm1.intercept_)#Intercepto
````

````python
ecuacion1 = f"y = {lm1.coef_[0]}x + {lm1.intercept_}"
print("La ecuacion de la linea de regresion es: ", ecuacion1)
````

### **Visualización de Resultados:**

````python
 #Definir tamaño
plt.figure(figsize=(6, 4))

#Crear el diagrama de dispersión con Línea de regresión
sns.set(style= "whitegrid")
#sns.regplot(x=x, y=y, scatter=True,  color="blue")
plt.scatter(x1, y1, color = 'blue', label = 'Datos')
plt.plot(x1, lm1.predict(x1), color = 'red', label = 'Linea de regresión')
plt.title('Diagrama de dispersión con linea de regresión ')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificacion')
plt.legend()
plt.grid(True)
plt.show()
````

### **Cálculo de la Correlación:**

````python
coef_correlacion1 = pearsonr(x1, y1)
print("El coeficiente de correlación de Pearson:", coef_correlacion1[0])
````

### **Conclusiones(Interpretacion de resultados):**


```
# Tiene formato de código
```


**¿Cuál es la interpretación del coeficiente de regresión en términos de la relación entre el tiempo de estudio y las calificaciones finales?**

La pendiente indica cuánto se espera que cambie la calificación por cada hora adicional de estudio. En este caso, como la pendiente es cercana a 0 o negativa, sugiere que no hay un impacto positivo fuerte del estudio sobre la calificación.

**¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y
dirección de la relación entre las variables**

El coeficiente de correlación de Pearson indica la fuerza y dirección de la relación. Si el valor es cercano a 0 (como en este caso, ~-0.057), hay muy poca o ninguna relación lineal entre las horas de estudio y la calificación.

**¿El modelo de regresión lineal es estadísticamente significativo?**

Dado que la correlación es baja y la pendiente no es considerable, el modelo no es estadísticamente significativo para predecir calificaciones con base en el tiempo de estudio.

**¿Qué tan confiable es el modelo de regresión para predecir las calificaciones finales de los estudiantes basado en el tiempo de estudio?**

El modelo no es confiable, ya que los resultados muestran una relación débil entre las variables.
