````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
from scipy.stats import pearsonr
````

# **Grupo 1**

### **Carga y Exploraci√≥n de Datos:**

````python
#Cargar el conjunto de datos del grupo 1 utilizando Python (pandas).
df1=pd.read_csv('/content/grupo1.csv')
df1.head()
````

````python
#Realizar una exploraci√≥n inicial de los datos (descripci√≥n estad√≠stica y visualizaci√≥n).
df1.info()
````

````python
df1.describe()
````

````python
#Variable independientes o variables predictorias
x1 = df1 [["HorasEstudio"]]
x1.head(10)
````

````python
#variable dependiente o variable objetiva a predecir
y1 = df1 [["Calificacion"]]
y1.head(10)
````

````python
#Generar un histograma de calificaciones
plt.figure(figsize=(6, 4))
sns.histplot(df1["Calificacion"], kde=True, bins=10, color='blue', edgecolor='black')
plt.title('Grupo 1 - Histograma de Calificaciones')
plt.xlabel('Calificaci√≥n')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
````

### **Modelo de Regresi√≥n Lineal:**

````python
#Entrenar un modelo de regresi√≥n lineal simple usando scikit-learn o statsmodels.
lm1 = LinearRegression()
lm1.fit(x1, y1)
````

````python
#Obtener la ecuaci√≥n de la l√≠nea de regresi√≥n (pendiente e intercepto).
print("Coeficiente de la pendiente (m):", lm1.coef_[0])#Pendiente
print("Intercepcion de la recta (b):", lm1.intercept_)#Intercepto
````

````python
ecuacion1 = f"y = {lm1.coef_[0]}x + {lm1.intercept_}"
print("La ecuacion de la linea de regresion es: ", ecuacion1)
````

### **Visualizaci√≥n de Resultados:**

````python
 #Definir tama√±o
plt.figure(figsize=(6, 4))

#Crear el diagrama de dispersi√≥n con L√≠nea de regresi√≥n
sns.set(style= "whitegrid")
#sns.regplot(x=x, y=y, scatter=True,  color="blue")
plt.scatter(x1, y1, color = 'blue', label = 'Datos')
plt.plot(x1, lm1.predict(x1), color = 'red', label = 'Linea de regresi√≥n')
plt.title('Diagrama de dispersi√≥n con linea de regresi√≥n ')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificacion')
plt.legend()
plt.grid(True)
plt.show()
````

### **C√°lculo de la Correlaci√≥n:**

````python
coef_correlacion1 = pearsonr(x1, y1)
print("El coeficiente de correlaci√≥n de Pearson:", coef_correlacion1[0])
````

### **Conclusiones(Interpretacion de resultados):**


```
# Tiene formato de c√≥digo
```


**¬øCu√°l es la interpretaci√≥n del coeficiente de regresi√≥n en t√©rminos de la relaci√≥n entre el tiempo de estudio y las calificaciones finales?**

La pendiente indica cu√°nto se espera que cambie la calificaci√≥n por cada hora adicional de estudio. En este caso, como la pendiente es cercana a 0 o negativa, sugiere que no hay un impacto positivo fuerte del estudio sobre la calificaci√≥n.

**¬øC√≥mo se interpreta el coeficiente de correlaci√≥n en t√©rminos de la fuerza y
direcci√≥n de la relaci√≥n entre las variables**

El coeficiente de correlaci√≥n de Pearson indica la fuerza y direcci√≥n de la relaci√≥n. Si el valor es cercano a 0 (como en este caso, ~-0.057), hay muy poca o ninguna relaci√≥n lineal entre las horas de estudio y la calificaci√≥n.

**¬øEl modelo de regresi√≥n lineal es estad√≠sticamente significativo?**

Dado que la correlaci√≥n es baja y la pendiente no es considerable, el modelo no es estad√≠sticamente significativo para predecir calificaciones con base en el tiempo de estudio.

**¬øQu√© tan confiable es el modelo de regresi√≥n para predecir las calificaciones finales de los estudiantes basado en el tiempo de estudio?**

El modelo no es confiable, ya que los resultados muestran una relaci√≥n d√©bil entre las variables.
